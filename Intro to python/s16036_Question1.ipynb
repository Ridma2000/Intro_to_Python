```python
#Question 1

import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

t = np.array([0,10,20,30,40,50])
N = np.array([1000, 853, 727, 619, 527, 448])
N0 = 1000 # Initial number of atoms

#least squares regression (linear curve fitting)
y = np.log(N)   #obtaining the natural logarithm of N,(transforming the exponential decay model into a linear form)
A = np.vstack([t, np.ones(len(t))]).T     #creating the matrix A with 2 columns with  t values and ones.
m, c = np.linalg.lstsq(A, y, rcond=None)[0] #solving A and y for m and c

#Printing the results 
print("a.)")
print("The slope of the line = -位 =", m) 
print("The intercept of the line =",c) 
print("The decay constant 位 = ", -m/N0 ,"per second")
print()

#Plotting data and the best fit line 
plt.scatter(t, y, color='g',label="Data points") #plotting the data points 
plt.plot(t, m*t + c, color='r', label="Fitting line") #plotting the best-fit line 
plt.xlabel("Time (s)") 
plt.ylabel("ln(N)") 
plt.title("The graph of ln(N) vs t") 
plt.legend()
plt.show() 

print("b.)")
# Predicted values
y_pred = m * t + c

# Residual sum of squares
ss_res = np.sum((y - y_pred) ** 2)

# Total sum of squares
ss_tot = np.sum((y - np.mean(y)) ** 2)

# R-squared
r_squared = 1 - (ss_res / ss_tot)

print("R-squared value :", r_squared)
```

```python
#Question 2

import numpy as np
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial

t = np.array([0, 2, 5, 9, 15])
h = np.array([0.0, 12.1, 65.5, 197.6, 499.0])

poly = lagrange(t, h)
print("a.)")
print('The Lagrange interpolation polynomial is: \n',poly)
print()

print("b.)")
# estimating velocity and acceleration when t = 7
velo = poly.deriv(1)(7)    #computes the first derivative of the polynomial and substitute 7 for it
acc = poly.deriv(2)(7)    #computes the second derivative of the polynomial and substitute 7 for it

print("velocity= ", velo,"ms-1")
print("acceleration= ", acc,"ms-2")
```
